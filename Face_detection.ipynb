```python
import cv2

# Load the Haar cascade file
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Start video capture (0 = default webcam)
cap=cv2.VideoCapture(0)

while True:
    # Read frame from webcam
    success,frame=cap.read()

    # convert to gray scale
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)
    print(faces)

    # Draw rectangle boxes arounf faces
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(143, 0, 255),3)
    
    # Display the frame
    cv2.imshow("Face Detection By Kevin Manjila",frame)

    # Press q to quit
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

# Release and close
cap.release()  # It disconnects from the camera device driver
cv2.destroyAllWindows()  # Closes all GUI windows created by OpenCV
```
